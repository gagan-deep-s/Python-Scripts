```python
import pandas as pd
```

```python
# Data Extract
# Import data from website on (The 50 largest companies by revenue in 2021 according to the Fortune India 500)
data=pd.read_html('https://en.wikipedia.org/wiki/List_of_largest_companies_in_India#2022_Forbes_list',index_col=False)
```

```python
# List top 5 data
data[1].head()
```

```python
# Assigning data to another data frame
top_comp=data[1]
```

```python
#Performing data cleanup
top=top_comp.loc[:, top_comp.columns != 'Revenue growth']
```

```python
# Replacing null data with character
top.fillna("Not Sure").head()
```

```python
# Replace index with Rank column
top.set_index('Rank',inplace=True)
```

```python
top.groupby('Name').mean()
```
